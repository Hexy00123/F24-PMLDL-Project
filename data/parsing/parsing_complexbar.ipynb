```python
import requests
from bs4 import BeautifulSoup
from time import time 
from tqdm.notebook import tqdm
```

```python
URL = 'https://complexbar.ru/recipe/'
PAGES_NUMBER = 57
CATEGORY_MAPPER = {
    'limonady': 'lemonade', 
    'kofeynye-napitki': 'coffee', 
    'milksheyki': 'milkshake', 
    'chai': 'hot_tea', 
    'kholodnye-chai': 'cold_tea', 
    'alkogolnye-koktejli': 'alcohol', 
    'smuzi': 'smoothie', 
    'punshi': 'punch', 
}
```

```python
drink_urls = []
for page_id in range(1, PAGES_NUMBER + 1): 
    page_url = URL + f'page-{page_id}'
    
    page_request = requests.get(page_url)
    print(page_url, page_request.status_code)

    soup = BeautifulSoup(page_request.content, "html.parser")
    drinks = soup.find('div', {'class': 'cmx-blog__items'}).find_all('a')

    for drink in drinks: 
        drink_urls.append(drink['href'])

drink_urls
```

```python
drinks = [] 

for drink_url in tqdm(drink_urls): 
    drink_request = requests.get(drink_url)

    category = CATEGORY_MAPPER[list(filter(lambda x: x, drink_url.split('/')))[-2]]
    print(category)
    soup = BeautifulSoup(drink_request.content, 'html.parser')

    drink_name = soup.find('h1').find('span').text

    drink_block = soup.find('div', {'class': 'cmx-blog-page-main'})

    image_link = drink_block.find('img')['src']
    image_ext = image_link.split('.')[-1]
    timestamp = str(time()).replace('.', '_')

    recipie_block = drink_block.find_all('div')[-1]
    recipie = recipie_block.text

    image_request = requests.get(image_link)

    with open(f'imgs/{timestamp}.{image_ext}', mode='wb') as file: 
        file.write(image_request.content)

    drinks.append({
        'name': drink_name, 
        'recipie': recipie,
        'category': category, 
        'image_name': f'{timestamp}.{image_ext}', 
        'image_ext': image_ext, 
        'image_url': image_link
    })

    # print(drinks[-1])
```

```python
import csv


with open('drink_recipies.csv', 'w', newline='') as csvfile:
    fieldnames = ['name', 'recipie', 'category', 'image_name', 'image_ext', 'image_url']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()
    writer.writerows(drinks)
```
